```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
usedcars=pd.read_json("E:/Python Programs/cars.json")
usedcars
```

```python
used_cars_model=usedcars.model.unique()
used_cars_model
```

```python
usedcars.model.value_counts()
```

```python
for x in used_cars_model:
    plt.scatter(x="price", y="mileage",data=usedcars[usedcars.model==x])
plt.xlabel("Price in Dollars")
plt.ylabel("Miles")
plt.title("Used Cars Price")
plt.show()
```

```python
plt.scatter(x="price", y="mileage",data=usedcars[usedcars.model=="SE"],c="Red", label="SE")
plt.scatter(x="price", y="mileage",data=usedcars[usedcars.model=="SES"],c="green", label="SES")
plt.scatter(x="price", y="mileage",data=usedcars[usedcars.model=="SEL"],c="black", label="SEL")
plt.xlabel("Price in Dollars")
plt.ylabel("Miles")
plt.legend(title="Car Models", loc="best",ncols=2)
plt.title("Used Cars Price")
plt.show()
```

```python
import numpy as np
import pandas as pd
import pickle

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('E:/Python Programs/titanic.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
```

```python
df.head()
```

## Step-1 Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Survived']), df['Survived'],test_size=0.2,random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
df.isnull().sum()
```

## Applying imputation

```python
si_age = SimpleImputer()
si_fare = SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

X_train_age = si_age.fit_transform(X_train[['Age']])
X_train_fare = si_fare.fit_transform(X_train[['Fare']])
X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])

X_test_age = si_age.fit_transform(X_test[['Age']])
X_test_fare = si_fare.fit_transform(X_test[['Fare']])
X_test_embarked = si_embarked.fit_transform(X_test[['Embarked']])
```

```python
X_train_age
```

## One Hot Encoding on Sex & Embarked in X_features

```python
ohe_sex = OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe_embarked = OneHotEncoder(sparse=False, handle_unknown='ignore')

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)

X_test_sex = ohe_sex.transform(X_test[['Sex']])
X_test_embarked = ohe_embarked.transform(X_test_embarked)
```

```python
X_train.head(2)
```

```python
X_train_sex
```

```python
X_train_embarked
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Fare','Embarked'])
```

```python
X_test_rem = X_test.drop(columns=['Sex','Age','Fare','Embarked'])
```

```python
X_train_rem.head(2)
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_fare,X_train_sex,X_train_embarked),axis=1)
```

```python
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_fare,X_test_sex,X_test_embarked),axis=1)
```

```python
len(X_test_sex)
```

```python
X_train_transformed.shape
```

```python
X_train_transformed
```

```python
X_test_transformed
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train_transformed,y_train)
```

```python
y_pred = dt.predict(X_test_transformed)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

## Exporting the model

```python
pickle.dump(ohe_sex,open('models/ohe_sex.pkl','wb'))
pickle.dump(ohe_embarked,open('models/ohe_embarked.pkl','wb'))
pickle.dump(dt,open('models/dt.pkl','wb'))
```

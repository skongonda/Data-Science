```python
# Handling Missing Data

# The data can be missing in 3 different forms and therefore we have 3 categories of Missing Data.
# They are  a) Missing Completely at Random (MCAR), b) Missing at Random (MAR), c) Missing Not at Random (MNAR)

# MCAR means the data has not been collected for some rows due to some reason, may be due to system failure, or may be program did not work, etc..,
# MAR means, the source from where your collecting your data did not provide some data in some rows, may be due their weight consciousness, or may be
# due to Age consciousness, or may be they did not want to give that data purposefully. So there is an assumption in MAR, that we can fill the missing
# values in a row, based on the other information available in the other columns of the row, i.e., there is a relation.
# MNAR means, the data has been removed purposefully and there is no much of a relation between the missing column values and the other columns. 


# Either you remove the Missing Values or Impute (i.e., Fill) them.
# There are 2 techniques for handiling missing values 1) Univariate 2) Multivariate

# In Univariate technique, we will focus only on one column while filling the missing values.
# Eg:- If you have multiple columns in your data, like Age, Height, Weight, etc.., and there are some missing values in Age column. So in 
# Univariate technique we will use only Age column to fill the missing values by may be taking the mean or median value of the existing values
# in the Age column.

# From Scikit learn module, using the class "SimpleImputer" we can fill the missing values. 
# For numerical columns, we will generally fill the missing values by mean(), or median(), or Random value, or by End of ditribution value.
# For categorical columns, we will generally fill the missing values by mode(), or you can mention the word "missing".

# Filling missing values
# from sklearn.impute import SimpleImputer
# si_age = SimpleImputer() ----> replaces missing values with mean() value.
# si_embarked = SimpleImputer(strategy = 'most_frequent') ----> replaces missing values with most frequent value.
# X_train_age = si_age.fit_transform(X_train[['Age']])  -------> this will give you a numpy array.
# X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']]) -------> this will give you a numpy array.
# X_test_age = si_age.transform(X_test[['Age']])  -------> this will give you a numpy array.
# X_test_embarked = si_embarked.transform(X_test[['Embarked']])  -------> this will give you a numpy array.

# Complete Case Analysis (CCA):- Complete Case Analysis (CCA), also called "list-wise deletion" of cases, consists in discarding
# observations (i.e., rows) where values in any variables (i.e., columns) are missing.
# Complete Case Analysis (CCA) literally means analyzing only those observations for which there is information in all of the variables in the dataset.

# Q) When to perform CCA?
# A) Our data is missing completely at random (MCAR). In case if you have a dataset where 1st 50 values are missing in any column or last
#    50 values are missing in any column, then you should not perform CCA i.e., should not remove those observations.
#    We will appply CCA only if the missing values are less than 5% in any given column. 
#    Also note that after perfroming CCA, i.e., if the data is MCAR and after performing CCA, the distibution of the data of those columns
#    should match with the new distribution of the data of the same column in the new dataset.

# We will replace the missing values with mean or median when the data is completely missing at random.
# But, if the data is not missing at random, and instead the data is missing in chunks, then we will use arbitrary value imputation technique.

# End of Ditribution technique is just an extension of Arbitrary value Imputation. But instead of choosing an arbitrary value, we will choose
# some value from the end of the distribution, either at left or right tail.
# For eg:- on whichever column you are applying End of Ditribution Imputation and if that column's data is normally distributed then we will
# use the value (mean + 3*StandDev) or otherwise (mean - 3*StandDev), because we all know that in 1 StandDev, there will be 34%+34%=68% 
# of the population, Similarly in 2 StandDev, there will be 95% of the population, and in 3 StandDev, there will be 99.7% of the population.
# So whatever values lie after 3 StandDev's are called Outliers and we use those values to Impute, if our data is normally distributed.
# check this image (https://www.google.com/search?sca_esv=838ff984a601b7f8&sxsrf=ADLYWIILN9GEDnKABjHHfFDtIw8bLCWS0Q:
#                  1723113673139&q=one+standard+deviation+percentage&tbm=isch&source=
#                  lnms&fbs=AEQNm0Aa4sjWe7Rqy32pFwRj0UkWERaHdBms7t-tHL1116ec0FnDIxrxgGhNFSZEtYqV91QHsjxJfdWVTuD0IrLFirCeLnWaVaCXk220-
#                  ef0J8Y_P9ub8y-nc4HAET9lqFGmo7CI5kB7gZ4uyIi2P8XL_hCFLNRNMklJaBc_s7xwJ3J7IGukfJo&sa=X&ved=
#                  2ahUKEwjYnZTTmuWHAxUkREEAHa1bJGIQ0pQJegQIDxAB&biw=1920&bih=911&dpr=1#imgrc=L2KaMW1dQhGbHM)
# But if the data in the column is skewed, then we will use IQR Proximity rule. Therefore the missing values will be imputed with the
# value of (Q1 - 1.5*IQR) or otherwise (Q1 + 1.5*IQR).
# Note IQR = Q3-Q1 i.e., 75th percentile - 25th percentile.

# So in short, arbitrary value imputation and end of the distribution imputation, both are same. The end goal for both the imputations is 
# to bring the importance of missing value to the surface. That's why you choose to fill an arbitrary value which is not present in your data or 
# you choose an extreme value which will lie after 3 StandDev's, called Outliers, to let our model know that these values are missing.

# Replacing missing values of categorical data:- Generally we will replace missing values of categorical data with the mode() which is
# also known as the most frequent occuring value.


# Random Imputation:-
# Let's say you have an Age Column and there are some missing values in it here and there. So in this case we will fill the missing values
# by Random Numbers, but these Random Numbers should be from the Age column itself from the existing Age values, by randomly selecting them.
# This technique can be applied to both, Numerical data and Categorical data.
# This technique is not readily available in Sklearn module and therefore we need to implement this by using Pandas.
# The benefit of this Random Imputation technique is that the data's distribution or Variance will be in intact i.e., will not change. 
# Where as in Mean/Meadian Imputation technique, the data's distribution migtht slighly change. 
# When you are working with Linear algorithms, Like Linear Regression or Logistic Regression, the data's disrubution shape and Variance 
# must be always intact. However this technique does not work good with Decsion Tree based algorithms because your will increase randomness
# in the data by doing so.
# The only small disadvantage in Random Imputation technique is that your co-variance with the other columns will slightly get's disturbed.
# Co-variance means relationship with other columns. And the reason for this is that you have introduced Randomness in your data.


# Missing Indicator Imputation:-
# In this technique we will completely create a new column for every column which has a missing data and you will keep only two values
# in this new column, either True or False.
# We will use MissingIndicator() class from sklearn.impute module

# Q) How can we automatically select the parameter values for any Imputation technique?
# A) We will use a technique called GridSearchCV.
#    In this technique, Sklearn will automatically try all the combinations and fetches you the best result. 


# In Multivariate technique, we will focus on all multiple columns at a time while filling the missing values.
# Eg:- If you have multiple columns in your data, like Age, Height, Weight, etc.., and there are some missing values in Age column. So in 
# Multivariate technique we will use all columns (i.e, Age, Height, Weight, etc..,) and rows to fill the missing values by may be taking the 
# mean or median values or some mathematical logical value which makes sense, of the all the existing values in all those columns.

# In Multivariate technique, we will use the classes KNN imputer, or Iterative imputer.

# KNN Imputer works on KNN algorithm. KNN's simple funda is that you are like your neighbours. Your entire behaviour, charecteristics,
# would be similar to your neighbour. So if there is any missing value in any column, then this missing value will be filled with that
# row which has the most similar behaviour, and charecteristics to your missing value row.
# And this similarity will be finded by using NaN-Euclidean distances.
# https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html
# We will use KNNImputer() class from sklearn.impute module

# NaN_euclidean_distances formula between two rows is:
# dist(x,y) = sqrt(weight * sq. distance from present coordinates)     where, weight = Total # of coordinates / # of present coordinates
# Eg:- the distance between [3, na, na, 6] and [1, na, 4, 5] is:
# sqrt(4/2 * ((3-1)**2 + (6-5)**2)
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.nan_euclidean_distances.html


# Multivariate Imputation by Chained Equations (MICE) or Iterative Imputation:- 
# We mostly use this MICE technique when the data is Missing at Random (MAR). i.e., using the other columns data we try and predict the missing values.
# MICE is always implemented only on Input columns but not on target column.
# Q) How do we implement MICE?
# A) Step 1:- Fill all the NaN values with mean() value of their respective columns.
#    Step 2:- Moving from left to right, you'll make the predicted mean() value in the 1st column as NaN again and feed it to the ML algorithm
#             of your choice. So a new value will be predicted. Please note that in this process your input data will be divided into train
#             and test data i.e.., the value where there is NaN or the missing column will be your y (Target) and existing values columns
#             will be your X (input).
#             After 1st column is finished, you move to 2nd column and repeat the same process until all columns are completed.
#             Now after all columns are completed and the missing values are filled once again, we will subtract the original mean() value data
#             with the new predicted value data to get a new data. Again taking the new data as your base data we will repeat the same steps moving
#             from left to right and again making the 1st column data NaN and predicting the value for that column. Once all columns are finished,
#             again your subtract your old data with new predicted data.
#             And this process will be repeated untill the predicted value becomes closer to the accurate value or close to 0.
```

```python
# There are two ways we can convert our numerical values into a categorical data. 1. Discretization (i.e., Binning)
# 2. Binarization

# "Discretization"/"Binning":- Binning is converting numerical values into a categorical data. It is the process of transforming continuous 
# variables into discrete variables by creating a set of contiguous intervals that span the range of the variable's values. Discretization
# is also called as Binning, where bin is an alternative name for interval.
# We use Discretization a) to handle outliers   b) to improve the value spread
# In Binning we have a) Unsupervised Binning which has 1) "Equal Width Binning" or "Uniform Binning"
# 2) "Equal Frequency Binning" or "Quantile Binning"  3) Kmeans Binning 
# b) Supervised Binning which has "Decision Tree Binning" (Rarely used).
# c) Custom Binning

# "Equal Width Binning" or "Uniform Binning":- In this binning we have to decide how many bins we want. 
# Formula:- (Max value - Min value)/bins 
# Imagine if max value is 100 and min value is 0 and you decide num of bins = 10 then (100-0)/10 = 10
# Therefore you will get 10 bins as (0-10), (10-20), (20-30),....(90-100).
# Generally in Equal Width Binning, your outliers are handled and there will be no change in the spread of the data.

# Eg:- You have a column called "weights" with values 50,55,61,65,75,77,79. Now in this case to convert these numerical values into categories,
# we can create a range such as 50-59, 60-69, 70-79 and these are called categories.
# Now you can map the numerical values in weights column according to their freequency i.e., 50-59 category has 2 values in weights, 
# 60-69 category has 2 values in weights, 70-79 category has 3 values in weights, and so on..
# https://www.youtube.com/watch?v=RfNfF1B5bfk
# https://www.youtube.com/watch?v=uBntfW8R7sw

# "Equal Frequency Binning" or "Quantile Binning":- In this binning also we have to decide how many bins we want. Suppose you decide you will
# need 10 bins, then each interval will contain 10% of total observations i.e., your 1st interval will start from 0, and reach to that  
# Age value where your 10th percentile resides i.e., (0-16) your 1st interval which is 10th percentile i.e., 10% of the population
# resides between age 0 to 16.
# Now 2nd interval will start at 16, and reach to that Age value where 20th percentile resides i.e., (16-20) your 2nd interval
# which is 20th percentile i.e. 20% of the population resides between age 0 to 20.
# Similarly 3rd interval will start at 20, and reach to that Age value where 30th percentile resides i.e., (20-22) your 3rd interval
# which is 30th percentile i.e. 30% of the population resides between age 0 to 22.
# In Equal Frequency Binning, the width of the interval will not be equal as you can see in the above example. I.e., (0-16) bin has 
# 16 width, (16-20) bin has 4 width, (20-22) bin has 2 width and so on...
# Equal Frequency Binning is most frequently used binning because it works well with the outliers and it will make your value data spread,
# Uniformly spread.

# "KMeans Binning":- This binning is specially used when you are working with clusters data. We can implement this binning using 
# sklearn.preprocessing module, which has "KBins Discretizer" class. When you create an object of this class, you pass 3 parameters a) num of bins
# b) Strategy = 'uniform' or 'quantile' or 'KMeans'  c) encoding = 'ordinal' or 'onehot' or 'onehot-dense'

# In Binarization binning, we usually convert a continuous value to a binary number (0 and 1). Eg:- Assume we have an Annual Income column,
# and you decide a threshold value say if your annual income is less than 6 lakhs per annum then you are not in taxable zone i.e., 0.
# But if your annual income is more than 6 lakhs per annum then you are in a taxable zone i.e., 1.
# So you have now converted your Annual Income column into a binary variable column, whether the person is taxable or not.
# The best use case for this Binarization binning is when you are working with Images and pixels. Generally in image processing we will
# have pixels from 0 to 255, and are color images. And if you want to convert them into black & white image, then you can keep a threshold of 127,
# saying if the pixel value is less than 127, then we will make it black i.e., 0. And if more than 127, then we will make it white i.e., 1.
# We can implement this binning using sklearn.preprocessing module, which has Binarizer class. When you create an object of this class,
# you pass 2 parameters a) threshold  b) copy = True or False.
```

```python
# Outliers

# Q) What are Outliers?
# A) Outliers are those observations or data points which are very different or has a different behaviour from the rest of the data points.
#    Imagine your data has outliers and you are training a machine learning model, then your ML model will get distracted or disturbed due to these
#    Outliers.

# Q) When are Outliers dangerous?
# A) Let's say you have Age column and there is one observation with 300 value. So in normal logical sense, a human cannot have a age of 300.
#    It might be either 30 or 3 or any other value. So in this case you simply remove this Outlier.
#    But there are some scenarios where Outliers are very important to system, like when you are working with Anamoly Detection Algorithms. Eg:- You
#    are working on credit card transactions data and you want to detect if there are any unusual transactions taking place. So using outliers we can 
#    detect such anamolies. But in this case if you remove Outliers then you cannot detect any anamolies. So in this case Outliers becomes
#    important for the data.

# Q) Effect of Outliers on ML algorithms?
# A) There are certain ML algorithms in which Outliers impact will lead to a poor performance, like Linear Regression, Logistic Regression, 
#    Adaboost, and Deep Learning. In all these algorithms, Outliers will have a deep impact on performance, because the common pattern in all
#    these algorithms is that we usually calculate weights.
#    But there are certain algorithms like Tree based algorithms such as Decision Tree, Random Forest, XGBoost, Gradient Boosting etc..,
#    Outliers does not create much of an impact.

# Q) How to treat Outliers?
# A) Let's say you found out that there are Outliers in your data. Then you either a) Trim the outliers i.e., remove those observations or
#    b) capping i.e., you keep a threshold on both the tails stating that the values less than 2 on left hand side will be treated as 2 only
#    and values greater than 90 on the right hand side will be treated as 90 only. This process is called capping.
#    c) "Discretization"/"Binning" by creating a set of contiguous intervals. Eg:- if say you have a value of 100 which is an outlier 
#    on your right hand side tail, then you you will merge that 100 value into 90-100 interval.

# Q) How to detect Outliers?
# A) If you want to know if there are any outliers in any of your columns, then we have 3 different techniques.
#    a) Suppose your column has a Normally distributed data or close to Normal distribution, then the values lesser than
#       (mean + 3*StandDev) and values greater than (mean - 3*StandDev) will be treated as Outliers, because we all know that 
#       in 1 StandDev, there will be (μ-1σ)+(μ+1σ) i.e., 34%+34% = 68% of the population, Similarly in 2 StandDev, there will be (μ-2σ)+(μ+2σ) i.e.,
#        13.6%+34%+34%+13.6% = 95% of the population, and in 3 StandDev, there will be (μ-3σ)+(μ+3σ) i.e.,
#       2.1%+13.6%+34%+34%+13.6%+2.1% = 99.7% of the population. So whatever values lie after 3 StandDev's are called Outliers.
#    b) Suppose your column does not have a Normally distributed data, instead the data is either right skewed or left skewed, then in this
#       case we will use IQR Proximity rule of BoxPlot. i.e., values lesser than minimum i.e., (Q1 - 1.5*IQR) and values greater than 
#       maximum i.e., (Q1 + 1.5*IQR) are treated as outliers. 
#       Note IQR = Q3-Q1 i.e., 75th percentile - 25th percentile.
#    c) Percentile approach i.e., If your column has any other distribution of the data, then we can keep a threshold on both the tail ends
#       that if the values on left hand side are less than 5 percentile and values on right hand side are greater than 90 percentile, then
#       those observations/values/data points will be treated as outliers. Threshold can be decided depending upon the problem statement.

# Q) what are the Techniques for Outlier Detection and Removal?
# A) a) ZScore treatment  b) IQR based filtering i.e., boxplot  c) Percentile approach  d) Winsorization


# ZScore Outlier Detection and Removal:- In this technique, on which ever column you are working to detect Outliers, that column should have 
# normally distributed data or close to normal distribution. Then you calculate transformed value for each X in that column by using  
# Formula Xi = (X - mean(X))/(Standard_Deviation) i.e., Xi = (X - μ)/ σ and then find out what is your lower limit i.e., (μ-3σ) 
# and upper limit i.e., (μ+3σ). So in this way you will able to find out what are the outliers in that column. Once the outliers are detected
# you can either Trim them i.e., (remove those observations) or you can perform capping i.e., replacing outliers with minimum value 
# on the left hand side and with maximum value on right hand side.

# Outlier Detection and Removal using IQR and BoxPlot:- In this technique, on which ever column you are performing Outlier detection, if that
# column does not have a Normally distributed data, instead the data is either right skewed or left skewed, then in this
# case we will use IQR Proximity rule of BoxPlot. i.e., values lesser than minimum i.e., (Q1 - 1.5*IQR) and values greater than 
# maximum i.e., (Q1 + 1.5*IQR) are treated as outliers.
```

```python
# PCA is generally an Unsupervised Machine Learning problem where we only have input and does not have output.
# The main goal of PCA is to reduce the curse of Dimensionality i.e., if you have multiple features in your data, and when you apply PCA, then 
# your features will get reduced. So in short PCA is a technique which can transform a higher dimensionality data to a lower dimensionality data 
# while keeping the essence of the data.
# Benefits of implementing PCA is that a) your model performance will improve by reducing the higher dimensionality data to a lower 
# dimensionality data which will make your algorithms execute much more faster and your computation will decrease. b) Second core benefit is that
# it will benefit your Visualization, because as a Human being we can only visualize 2D data or max to max we can visualize data in 3D. But if we 
# have 10 columns of data then it will become very complex to visualize the data in 10D.

# Q) How do we select which column should i keep and which column should i not keep among all the existing columns in our data?
# A) If you are working with a data where you have no idea which column is better between all the existing columns, then plot the scatter plot 
#    for every 2 columns and find out the projection of those 2 columns. On whichever axis you get more spread of the data (i.e., the variance)
#    then that column will become important for you and this is how feature selection works.
#    Eg:- Let's say you have 3 columns of data "Num of Rooms", "Num of Grocery Shops", "House Price". So if anyone asks you to eliminate one column 
#    out of 2 input columns, then we will generally plot a scatter plot for "Num of Rooms" and "Num of Grocery Shops" and check out the projection
#    of the data. Therefore whichever axis you see there is more spread, you keep that column and eliminate the other column i.e., d > d' where 
#    d is your x-axis (Num of Rooms) and d' is your y-axis (Num of grocery shops). 
#    However note that there will be some scenarios where this feature selection technique will not work and you should be going for feature 
#    extraction technique in such cases.
#    Eg:-  Let's say you have 3 columns of data "Num of Rooms", "Num of Washrooms", "House Price". So if anyone asks you to eliminate one column 
#    out of 2 input columns, then we cannot use feature selection technique of projecting the data and check out for the spread, because both the
#    axis's will have equal spread i.e., your House Price will always vary according to "Num of Rooms", and "Num of Washrooms".
#    i.e., More Num of Rooms means more Num of Washrooms which will lead to higher price of the house. So here d ≈ d'. So therefore in this situation
#    we should go for Feature Extraction technique. So what we will do is, instead of keeping 2 input columns "Num of Rooms", "Num of Washrooms", 
#    we can simply introduce a new column called "Size of the Flat" which will keep the essence of the data intact, because generally in a logical
#    sense we all know that if we have more num of rooms, means more num of washrooms and indirectly the size of the house will also be bigger. So 
#    short we have completely changed the 2 input columns to 1 input column i.e., transformed 2D data to 1D data.

# Q) What exactly does PCA do differently?
# A) Lets imagine we have 3 columns of data "Num of Rooms", "Num of Washrooms", "House Price". We will plot the 2 input columns "Num of Rooms",
#    "Num of Washrooms" in a scatter plot. If the data is linearly growing then we cannot differentiate between the variance on both the axis's.
#    Therefore what PCA does is, that it will shift the cordinate axis to the best fitting line. So in this way, on one axis may be x-axis or
#    on y-axis there will be a bigger spread on the other it will have a smaller spread. So the axis which has bigger spread will become PCA1 column
#    and the axis where the spread is smaller, will become PCA2 column. Therefore d > d'. 

# Q) Why is Variance important?
# A) Basically Variance will help you understand the spread of your data.
#    Imagine you have 3 data points -5, 0, 5. So if you find out mean of these 3 data points, it will be 0 because -5+0+5/3 = 0.
#    Similarly if you have another 3 data points -10, 0, 10. So if you find out mean of these 3 data points, it will be 0 because -10+0+10/3 = 0.
#    But notcie that the spread between these 2 datasets. First dataset has a lower spread and Second datset has higher spread.
#    So if you remove variance of these 2 datasets you will find out that (-5)**2 +(0)**2 +(5)**2 / 3 = 25+0+25/3 = 50/3
#    and  (-10)**2 +(0)**2 +(10)**2 / 3 = 100+0+100/3 = 200/3.
#    Formula ∑i=1 to n (Xi - x̄)**2 / n

# PCA step by step:- https://www.youtube.com/watch?v=tXXnxjj2wM4
# First you bring your data to mean center by using StandardScaler() function.
# In second step we will find covariance matrix by using np.cov() function.
# Third step is you find eigen values and eigen vectors for the covariance matrix by using np.linalg.eig() function.
# Fourth step is to select top two eigen vectors, which will be our PC1 and PC2.
# Fifth step we will transform our higher dimension data (i.e., n columns/X features) to two columns data that'll be our PC1 & PC2
# by using np.dot() function. Note - Eigen vectors should be transposed here before applying dot() function.
```

```python
# In Regression algorithms, our Target variable (i.e., y) will be a contious data. (eg's:- 1. How much
# loan amount should be approved  2. House price prediction  3. Cars price prediction).
# Types of different Regression algorithms are 1. Linear 2. Logistic 3. Lasso 4. Ridge etc..,
# Regression is nothing but a mathematical equation (ax+by+c=0) which explains the variation of the dependent variable
# (i.e., y) with respect to the the series of independent variables (i.e., x)
# In mathematics, a simple linear equation will be y=mx+c where "m" is the slope and "c" is 
# the intercept (can be x-intercept or y-intercept).
# where as in machine learning, a simple linear equation will be y=wx+b where "w" is the weight and "b" is the bias.
# So if you give x & y, the machine is going to tell you what is "w" and "b"
# If you book a cab, minimum price is "6-RS" and for every 1 KM you travel, you'll have to pay "3-RS"
# then the euation will be something like "y=3x+6"  OR  "Amount = 3 * Distancein KM's + 6RS"
# If you have a single independent varaible (i.e., only one X attribute) then the equation will be simple y=wx+b.
# But if you have multiple independent variables (i.e., multiple X attributes) then the equation will be something like
# y = w1.x1 + w2.x2 +.....+ wn.xn + b (here machine will try to learn all the weights w1,w2,...wn and bias b) and it's
# called Multiple Linear Regression
# y= ∑i=1 to n WiXi + b (where "n" is no:- of dimensions or we can say no:- of features)

# In Regression model, Accuracy score (also called as "R-Squared score" or "Co-effiecient of Determination") is calculated 
# in a different method. i.e., How much percentage of variation, the regression line was able to 
# explain. (basically variation between x & y or variation between your dependent and independent variable) 

# To calculate R-squared value, we need to do two calculations. "SSEReg" and "SSETotal".
# SSEReg is called as Sum of the squared error with Regression line. i.e., Σi=1 to n(y_actual values - y_predicted values)**2
# SSETotal is called as Sum of the squared error with Total. i.e., Σi=1 to n(y_actual values - mean value of the actual values)**2
# Therefore if you divide "SSEReg / SSETotal", it will give you ==> "How much percentage of variation which
# was unable to be explained by our Regression line".
# And your R squared value will be calculated as "1 - (SSEReg/SSETotal)" ==> which will give you "How much
# percentage of variation which was able to be explained by our Regression line".
# Note - Higher R-Squared value means it's a good model, probably profitable.
# Note - Lower R-Squared value means it's a bad model, probably dangerous.

# Proving Σ(xi-x̄) = 0 ==> "x1,x2,x3,...xn" and "mean=x̄"
# ==> (x1-x̄)+(x2-x̄)+(x3-x̄)+....+(xn-x̄)
# ==> (x1+x2+x3+...+xn)-n*x̄
# ==> (x1+x2+x3+...+xn)-n*((x1+x2+x3+...+xn)/n)
# ==> 0

# Residual = y_actual_values - y_predicted_values

# if "y_actual_value > y_predicted_value" then we call it as a +ve Residual
# if "y_actual_value < y_predicted_value" then we call it as a -ve Residual

# Best Fitting Line is the summation of all the +ve & -ve Residuals
# i.e., Σ Residuals = 0

# Q) How to find weights & bias?
# A) w, b = model.coef_, model.intercept_

# Q) How to calculate y_predict manually?
# A) y_predict = X.dot(w)+b
# A) y_predict = X.age*w[0] + X.smoker_yes*w[1] + X.bmi_30_smoker*w[2] + X.children_yes*w[3] + b
```

```python
# There are 2 ways for model evaluation.

# a) Metrics based evaluation:- There are some different metrics such as "accuracy score", "Balanced accuracy score",
# "Cohen Kappa score", "Precision", "Recall", "F1-measure," "TPR,FPR,TNR,FNR", "ROC & AUC score", "Sensitivity",
# "Specificity", etc.., 

# b) Sampling based evaluation:- There are different types of sample based techniques, such as "Holdout method",
# "K-fold cross validation", "Stratified K-fold cross validation", "leave-one-out", "leave-p-out",
# "Bootstrap Sampling", etc..
```

```python
# Accuracy score, Balanced accuracy score, and Cohen Kappa score are giving the overall performance of a 
# particular model, that means it includes all the classes that is getting predicted.
# But generally when it comes to reality, we are having a "Class of Interest".
# i.e., Mainly we are trying to apply Machine Learning algorithm to predict a particular class. For eg:- In a 
# Loan example of banking application, our Target is to find "whether a customer is a Defaulter". So "Defaulter" is 
# our class of interest.
# So in every classification problem, one class will be our "class of interest", which we are going to call it as
# "a +ve class". 
# And the remaining classes are going to be "Not the classes of Interest" whcih we are going to call it as " a -ve class".

# True Positive:- A postive correctly classified as positive
# True Negative:- A negative correctly classified as negative
# False Positive:- A negative wrongly classified as positive (also called as Type-1 errors)
# False Negative:- A postive wrongly classified as negative (also called as Type-2 errors)

# Total Actual Positives = TP + FN
# Total Actual Negatives = Fp + TN
# Total Predicted Positives = TP + FP
# Total Predicted Negatives = FN + TN
# Accuracy score = TP + TN / TP + FN + TN + FP
```

```python
# Cohen Kappa score:- How much agreement between actual value and predicted value.
# If the Cohen Kappa score is "> 0.8", then it is a perfect agreement.
# If the Cohen Kappa score is in between "0.6 to 0.8", then it is a Good agreement.
# If the Cohen Kappa score is in between "0.4 to 0.6", then it is a above average agreement.
# If the Cohen Kappa score is in between "0.2 to 0.4", then it is a average agreement.
# If the Cohen Kappa score is in between "0 to 0.2", then it is a poor agreement.

# Formula = (Probability(actual value) - Probability(expected value)) / 1 - Probability(expected value)
# where "Probability(actual value)" is just your accuracy score.
# "Probability(expected value)" is How many total apple actual values are there in test * How many total apples model predicted

# Q) What does it mean to say "Kappa score=0" or When will you get "Kappa score=0" ?
# A) Whenever TPR=0 or FPR=0, or any one of them is zero, then automatically your kappa score will become zero.
```

```python
# Precision score:- Your precision score is nothing but, how many you attempted and how many are correct out of them.
# Eg:- if your precision score is 0.82 then we can say, "Out of the total postives it predicted, 82% are positives".
# Formula = TP / TP + FP
# or in Confusion matrix we can say "TP / Column Total" (provided "predicted values" 
# are in columns and "actual values" are in rows.)

# Recall score:- Recall score is nothing but, how many you are expected to attempt and how many
# you have attempted and how many are correct out of them.
# Eg:- if your recall score is 0.84 then we can say, "Out of the total postives i have given you, only 84% you were
# able to identify".
# Formula = TP / TP + FN
# or in Confusion matrix we can say "TP / Row Total" (provided "predicted values" 
# are in columns and "actual values" are in rows.)

# Eg:- Total 5 questions given. You attempted 4 questions & 3 are correct.
# Therefore, your Precision score is 3/4 = 75%
# your Recall score is 3/5 = 60% 

# Based on the problem statement, if you think FP is more harmfull, then you Precision score should be higher.
# if you think FN is more harmfull, then you Recall score should be higher.

# In case if you feel, you FP and FN, both are harmfull in any scenario, then you can decide by using F1-measure score.
# F1-score will combine both the values, i.e., your Precision score and Recall score into one value.
# F1-score is nothing but, Harmonic mean of Precision score and Recall score.
# Formula ==> 1/F = (1/P + 1/R)*1/2
# ==> 1/F = (P+R/P*R)*1/2
# ==> F = (2*P*R/P+R)

# If F > 0.8, then this means that your model is good.
```

```python
# TPR (True Positivity Rate):- The ability of the model in identifying the positive samples as positive samples.
# (Out of the total actual positives, how many are correctly classified as positives.)
# Formula = TP / TP + FN ==> also called as Recall / Senitivity / Hit Ratio

# FNR (False Negativity Rate):- The ability of the model in identifying the positive samples as negative samples.
# Formula = FN / TP + FN

# Therfore, TPR + FNR = 1   (or)    FNR = 1 - TPR     (or)    TPR = 1 - FNR 

# FPR (False Positivity Rate):- The ability of the model in identifying the negative samples as positive samples.
# (The ratio of the negative samples, which are wrongly classified as positive.)
# Formula = FP / FP + TN

# TNR (True Negativity Rate):- The ability of the model in identifying the negative samples as negative samples.
# (The ratio of the negative samples, which are correctly classified as negatives.)
# Formula = TN / TN + FP ==> also called as Specificity

# # Therfore, FPR + TNR = 1   (or)    FPR = 1 - TNR     (or)    TNR = 1 - FPR

# Q) How to decide whether a model is good model or bad model?
# A) A model which is well differentiating between positives and negatives is always a good model.
#    i.e., when TPR = 1 and FPR = 0, it is a good model.
#    A good model should always have a higher TPR and a lower FPR.
#    A worst model is when when TPR = 0 and FPR = 1.

# Q) What does it mean if TPR=0 & FPR=0 ?
# A) TPR=0 means, All positive samples are predicted as negatives, which means FNR = 1.
#    FPR=0 means, All negative samples are predicted as negatives, which means TNR = 1.
#    This means, all samples are predicted as negatives.
#    A model which is having a lower TPR and FPR is not a good model.

# Q) What does it mean if TPR=1 & FPR=1 ?
# A) TPR=1 means, All positive samples are predicted as positives, which means FNR = 0.
#    FPR=1 means, All negative samples are predicted as positives, which means TNR = 0.
#    This means, all samples are predicted as positives.
#    Not a good model.
```

```python
# ROC curve (Reciever Operating Charecteristic curve):-
# We will use the ROC curve to check which model is well differentiating between positives & negatives. 

# ROC curve is plotted by visualizing FPR & TPR values.
# FPR will always be on the x-axis and TPR will always be on the y-axis.
# Threshold values will be from [2 to 0].
# When Threshold value is 0, then TPR=1 & FPR=1.
# When Threshold value is >=2, then TPR=0 & FPR=0.
# The model which has highest AUC (Area under the curve) will be a good model.
# If AUC=1, then TPR=1 and FPR=0 at a particular threshold value.
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import (train_test_split, KFold, LeaveOneOut, LeavePOut, RepeatedKFold, RepeatedStratifiedKFold,
                                     StratifiedKFold)
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score, accuracy_score
```

```python
fruits=pd.read_csv("E:/Python Programs/fruits.csv")
X = fruits[["sphericity", "weight"]]
y = fruits["label"]
```

```python
kf1 = KFold(n_splits=5, shuffle=False)
kf1.split(X)
```

```python
for ix_train, ix_test in kf1.split(X):
    print(ix_train)
    print(ix_test)
    print()
```

```python
kf2 = KFold(n_splits=5, shuffle=True)
kf2.split(X)
```

```python
for ix_train, ix_test in kf2.split(X):
    print(ix_train)
    print(ix_test)
    print()
```

```python
model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
for ix_train, ix_test in kf2.split(X):
    X_train = X.iloc[ix_train]
    X_test = X.iloc[ix_test]
    y_train = y.iloc[ix_train]
    y_test = y.iloc[ix_test]
    model.fit(X_train, y_train)
    accuracy.append(model.score(X_test, y_test))
print(accuracy)
```

```python
np.mean(accuracy)
```

```python
loo = LeaveOneOut()
for ix_train, ix_test in loo.split(X):
    print(ix_train)
    print(ix_test)
    print()
```

```python
loo = LeaveOneOut()
dt_model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
for ix_train, ix_test in loo.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    dt_model.fit(X_train, y_train)
    dt_y_predict = dt_model.predict(X_test)
    accuracy.append(dt_model.score(X_test, y_test))
print(accuracy)
print(np.mean(accuracy))
```

```python
knn_model = KNeighborsClassifier()
accuracy = []
for ix_train, ix_test in loo.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    knn_model.fit(X_train, y_train)
    knn_y_predict = knn_model.predict(X_test)
    accuracy.append(knn_model.score(X_test, y_test))
print(accuracy)
print(np.mean(accuracy))
```

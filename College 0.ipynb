```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from pylab import *
```

```python
# Random data that shows a roughly linear relationship between page speed and amount purchased
```

```python
pagespeeds=np.random.normal(3.0, 1.0, 1000)
purchaseamount=100-(pagespeeds + np.random.normal(0,0.1,1000))*3
```

```python
pagespeeds[0:5], purchaseamount[0:5]
```

```python
sns.scatterplot(x=pagespeeds, y=purchaseamount)
plt.plot([0,0],[98,100],c="black")
plt.xlabel("Page Speed")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
# as we have only 2 features, we can keep it simple and just use "scipy.stats.linregress" function.
```

```python
slope, intercept, rvalue, pvalue, stderr = stats.linregress(pagespeeds, purchaseamount)
```

```python
slope, intercept, rvalue, pvalue, stderr
```

```python
# Now surprisinly, our "r-value squared" shows a really good fit. 
```

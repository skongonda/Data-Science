```python
# Out of 60 fruits, making 1770 combinations of 2 fruits for testing i.e., 60-C-2 ==> 1770 

lpo = LeavePOut(p=2)
model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
start_time = datetime.now()
for ix_train, ix_test in lpo.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    model.fit(X_train, y_train)
    accuracy.append(model.score(X_test, y_test))
end_time = datetime.now()
print("Time taken to execute: ", end_time - start_time)
print(accuracy)
```

```python
# 4 (n_splits=4) samples repeated for 5 (n_repeats=5) times

rkf = RepeatedKFold(n_splits=4, n_repeats=5)
model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
start_time = datetime.now()
for ix_train, ix_test in rkf.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    model.fit(X_train, y_train)
    accuracy.append(model.score(X_test, y_test))
end_time = datetime.now()
print("Time taken to execute: ", end_time - start_time)
print(accuracy)
```

```python
len(accuracy)
```

```python
# taking 5 samples 5 times

spf = ShuffleSplit(n_splits=5)
model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
start_time = datetime.now()
for ix_train, ix_test in spf.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    model.fit(X_train, y_train)
    accuracy.append(model.score(X_test, y_test))
end_time = datetime.now()
print("Time taken to execute: ", end_time - start_time)
print(accuracy)
```

```python
len(accuracy)
```

```python
for ix_train, ix_test in spf.split(X):
    print("Training Indexes")
    print(ix_train)
    print("Testing Indexes")
    print(ix_test)
    print()
```

```python
# taking random 20 samples (test_size=20) and repeat it 5 times (n_splits=5)

spf_1 = ShuffleSplit(n_splits=5, test_size=20)
model = DecisionTreeClassifier(criterion='entropy')
accuracy = []
start_time = datetime.now()
for ix_train, ix_test in spf_1.split(X):
    X_train, X_test, y_train, y_test = X.iloc[ix_train], X.iloc[ix_test], y.iloc[ix_train], y.iloc[ix_test]
    model.fit(X_train, y_train)
    accuracy.append(model.score(X_test, y_test))
end_time = datetime.now()
print("Time taken to execute: ", end_time - start_time)
print(accuracy)
len(accuracy)
```

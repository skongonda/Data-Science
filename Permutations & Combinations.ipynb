```python
from itertools import permutations
from itertools import combinations
```

```python
numbers = [x for x in range(1,11)]
```

```python
numbers
```

```python
perm = permutations(numbers, r=3)
```

```python
for i in perm:
    print(i)
```

```python
comb = combinations(numbers, r=3)
```

```python
for i in comb:
    print(i)
```

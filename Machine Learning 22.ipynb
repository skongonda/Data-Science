```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
fruits=pd.read_csv("E:/Python Programs/fruits.csv")
X = fruits[["sphericity", "weight"]]
y = fruits["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=10)
```

```python
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
# Sampling with replacement, and hence samples are repeated

fruits_bootstrap = fruits.sample(n=60, replace=True)
fruits_bootstrap
```

```python
# unique samples

len(np.unique(fruits["sphericity"]))
```

```python
fruits_bootstrap = fruits.sample(n=60, replace=True).reset_index(drop=True)
fruits_bootstrap.shape
```

```python
# another way to find unique samples

fruits_bootstrap.drop_duplicates().shape
```

```python
# again splitting the data from bootstrap sampling

X = fruits_bootstrap[["sphericity", "weight"]]
y = fruits_bootstrap["label"]

X_train, X_test, y_train, y_test = train_test_split(X,y)
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

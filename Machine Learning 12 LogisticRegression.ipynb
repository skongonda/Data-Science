```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, minmax_scale, MinMaxScaler
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, r2_score
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
X = iris[['petal_length', 'petal_width']]
y = iris['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=10)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
dt_model = DecisionTreeClassifier(criterion='entropy')
dt_model.fit(X_train, y_train)
dt_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, dt_model.predict(X_test))
plt.show()
```

```python
knn_model = KNeighborsClassifier(weights='distance')
knn_model.fit(X_train, y_train)
knn_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, knn_model.predict(X_test))
plt.show()
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_model.score(X_test, y_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, lr_model.predict(X_test))
plt.show()
```

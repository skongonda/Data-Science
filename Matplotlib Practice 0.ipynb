```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# mainly visualizations will be categorized into three types
# 1. Relational Plots 2. Distribution Plots 3. Categorical Plots


## Relational Plots:- 

# Relational Plots are mainly used to find out how two variables are related. (Eg:- Age & Height). To use the relational
# plot, compulsory two variables (x & y axis) are required and it's always should be continuous data (Continuous vs Continuous), 
# and we don't use categorical data at all. 
# There are two types of relational plots, mainly "Scatter plots" and "Line plots".
# scatter plots and line plots will show you the actual data points.
# When we want to check where the most number of data points are lying, then we will use "scatter plots".
# When we want to check the growth over a time period, then we use "line charts/plots". For line plot, it's not required to have
# 2 variables compulsory. We can even use one variable and that variable will be taken into y-axis and whatever the index is,  
# it will be automatically taken into the x-axis.

# "Rel Plot" is also one of the relational plot just like scatter plot and lineplot.


# Categorical Plots:-

# "Bar charts" - Whenever we want to visualize categorical data, then we use bar charts mostly. To use bar charts, we should have
# two variables x & y axis (Either "x and height" OR "y and width"). 
# Generally in bar plots, one axis (mostly "x-axis") will be a categorical data and one axis may be a continuous data. 
# But we do not plot categorical data on both axis. Either it must be a "Categorical vs Continuous" or 
# "Continuous vs Continuous".
# Bar charts will only show you the quantity. Either it can be, for each categorical, what is the count. Or for each
# categorical, what is the average etc..,

# "Box plots" are not used to visualize actual data points, but instead they are used to visualize the quantiles. It can also be
# used to plot the outliers in the data points. In boxplot, we use single variable only and we cannot use mutiple variables.
# The data you specify, will come on y-axis, but by notation we specify "x".
# IQR(inter quantile range) = Q3-Q1
# Formula to calculate outlier = Q1-1.5*IQR is the lower cut off value.
# Formula to calculate outlier = Q3+1.5*IQR is the higher cut off value.

# "Count Plots" are just like Bar plots. In "Count plot" we should either give x or y, but should not give both of them as it
# will not work or it may give error. One axis will be categorical, which you you will be specifying and the 
# other axis will be automatcially treated as the count of those categorical values.

# "Cat Plots" are also just like Bar plots. In "Cat plot" we should either give x or y, along with specifying the parameter
# "kind=count", but should not give both of them as it will not work or it may give error.
# Only at the time of using parameter "kind=bar" or "kind=swarm" or "kind=strip", we should give both "x and y"

# "Swarm Plots" are used to see the overlapping data points. Basically in scatter plot the overlapped data points are not
# visible, where as if you choose "swarm plot", then the overlapped data points will be visible.
# In "swarm plots", we will have x & y axis and we can give either "x & y", or "only x" or "only y".
# In "swarm plots" we can either use "Categorical vs Continuous" or "Categorical vs Categorical" or can be 
# just "Categorical" with different parameters such as "Hue", and "Size".

# Similar to "swarm plot" we have "Strip plot". It is basically used in larger data set.


# "Pie charts" does not represent actual values, but instead it represents only the proportion. In pie charts we don't have
# x and y axis, but by notation we specify "x".


# Distribution plots:-

# "Hist Plots", "kde plot", "ecdf plot" and "rug plot" are the examples of Distribution plots.
# "Hist plots" are basically just like count plot. In Count plots we will use categorical data, where as in Hist plots
# we will use continuous data. 
  
```

```python
angles = np.arange(0,1441)
sine = np.sin(np.radians(angles))
cosine = np.cos(np.radians(angles))
```

```python
plt.axes(facecolor='black')
plt.plot(angles,sine,color='red',ls="--")
plt.plot(angles,cosine,color='yellow',ls="-.")
plt.show()
```

```python
# x and y values shape should match, otherwise we will get error. eg:- here x shape is 12 months 
# and y shape is sales of 12 months between 1 and 500 dollars

plt.bar(x=np.arange(1,13),height=np.random.randint(1,500,12))
plt.show()
```

```python
# x and y values shape should match, otherwise we will get error. eg:- here x shape is 12 months 
# and y shape is sales of 12 months between 1 and 500 dollars

plt.barh(y=np.arange(1,13),width=np.random.randint(1,500,12))
plt.show()
```

```python
x=np.linspace(-6,6,50)
y=x**4
z=x**3
plt.axes(facecolor="black")
plt.scatter(x,y,c="green")
plt.scatter(x,z,c="yellow")
plt.show()
```

```python
plt.pie(x=np.random.randint(50,500,6),explode=[0,0.1,0,0.1,0,0.1],labels=[1,2,3,4,5,6])
plt.show()
```

```python
iris=pd.read_csv("E:/Python Programs/iris.csv",header=1,names=["sepal_length_cm","sepal_width_cm","petal_length_cm","petal_width_cm","species"])
iris
```

```python
iris["species"].value_counts()
```

```python
# Renaming values of species column by removing "Iris-" using "lambda" and "split()"

iris.species = iris.species.apply(lambda x: x.split("-")[1])
iris
```

```python
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris)
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.scatter(x=iris.sepal_length_cm, y=iris.petal_length_cm)
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.axes(facecolor="black")
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris[iris["species"]=="setosa"], marker="D",color=[[1,0,1]])
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Setosa Sepal Length vs Setosa Petal Length")
plt.show()
```

```python
plt.axes(facecolor="black")
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris, marker="^",color=np.array([[217,128,97]])/255)
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.axes(facecolor="black")
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris, marker="*",color=np.array([[217,128,97]])/255)
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.axes(facecolor="black")
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris, marker="*",color="#EA3680")
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.axes(facecolor="black")
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris[iris["species"]=="setosa"])
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris[iris["species"]=="versicolor"])
plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris[iris["species"]=="virginica"])
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
spe = iris.species.unique()
for s in spe:
    plt.scatter(x="sepal_length_cm", y="petal_length_cm", data=iris[iris.species==s])
plt.xlabel("Sepal Length in CM's")
plt.ylabel("Petal Length in CM's")
plt.title("Sepal Length vs Petal Length")
plt.show()
```
